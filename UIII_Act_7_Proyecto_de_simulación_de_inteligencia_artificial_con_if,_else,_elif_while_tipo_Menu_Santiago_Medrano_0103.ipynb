#### Este programa simula un sistema de IA de Seguridad Urbana diseñado para procesar reportes de la policía y gestionar una alarma de emergencia

```python
# --- SIMULADOR DE IA: SISTEMA DE SEGURIDAD INTEGRADO ---

print("INSTALANDO IA DE SEGURIDAD CIUDADANA...")
print("---------------------------------------")

sistema_activo = "si"

# Ciclo principal para mantener el programa interactivo [4, 5]
while sistema_activo == "si" or sistema_activo == "SI":
    print("\n--- MENÚ DE CONTROL IA ---")
    print("1. Reporte de Policía")
    print("2. Alarma de Emergencia")
    print("3. Estado del Sistema")
    print("4. Salir")

    opcion = input("Seleccione una opción (1-4): ")

    # Lógica para la función de Policía [2, 6]
    if opcion == "1":
        print("\n[MÓDULO POLICÍA SELECCIONADO]")
        tipo_incidente = input("Tipo de incidente (Robo / Tráfico): ").lower()
        gravedad = int(input("Nivel de gravedad (1-10): "))

        # Uso de operadores lógicos (and) y relacionales (>) [3, 7]
        if tipo_incidente == "robo" and gravedad >= 7:
            print("IA-POLICÍA: Alerta crítica detectada. Desplegando unidades tácticas.")
        elif tipo_incidente == "robo" or tipo_incidente == "trafico":
            print("IA-POLICÍA: Reporte registrado. Enviando patrulla de zona.")
        else:
            print("IA-POLICÍA: Tipo de incidente no reconocido para respuesta automática.")

    # Lógica para la Alarma de Emergencia [8, 9]
    elif opcion == "2":
        print("\n[SISTEMA DE ALARMA DE EMERGENCIA]")
        nivel_ruido = int(input("Ingrese nivel de decibelios detectado (0-120): "))
        movimiento = input("¿Se detecta movimiento inusual? (si/no): ").lower()

        # Uso de operadores lógicos (or) para activar la alarma [3]
        if nivel_ruido > 90 or movimiento == "si":
            print("!!! ALARMA ACTIVADA !!! Protocolo de evacuación iniciado.")

            # Cálculo aritmético simple de riesgo [10]
            riesgo = (nivel_ruido * 0.5) + 10
            print(f"Cálculo de riesgo estimado: {riesgo}%")
        else:
            print("Alarma en modo espera. No se detectan amenazas.")

    # Opción de Estado del Sistema
    elif opcion == "3":
        print("\n--- ESTADO ACTUAL ---")
        print("Sensores: Operativos")
        print("Conexión con Central: Estable")
        print("IA Aprendiendo: Si")

    # Opción para salir del programa y romper el ciclo [4, 11]
    elif opcion == "4":
        sistema_activo = "no"
        print("Cerrando sistema de IA... Seguridad desactivada.")

    # Validación de entrada del menú [12]
    else:
        print("Opción no válida. Por favor, intente de nuevo.")

# Mensaje final fuera del ciclo [13]
print("\n--- PROGRAMA FINALIZADO ---")
```
